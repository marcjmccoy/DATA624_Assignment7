```python
#import dependencies
import dash_bootstrap_components as dbc
from dash.dependencies import Output, Input
from dash import html
from dash import dcc
from jupyter_dash import JupyterDash
import pandas as pd
import numpy as np
import plotly.express as px
```

# Wrangling

```python
data = pd.read_csv("13100096.csv")
data = data[data.GEO != "Canada (excluding territories)"]
data
```

```python
data_pivot = data.pivot( index= ['GEO', 'Age group', 'Sex', 'Characteristics', 'REF_DATE'],
                        columns= 'Indicators', values='VALUE').reset_index()
data_pivot
```

```python
provinces = pd.read_csv("provinces.csv")

provinces = provinces[provinces.Provinces != "Yukon"]
provinces = provinces[provinces.Provinces != "Northwest Territories"]
provinces = provinces[provinces.Provinces != "Nunavut"]

provinces = provinces.rename(columns={"Provinces": "GEO"})

provinces
```

```python
healthcan_df = pd.merge(data_pivot, provinces, left_on='GEO', right_on='GEO', how='left')
healthcan_df
```

```python
app = JupyterDash(__name__, 
                external_stylesheets=[dbc.themes.BOOTSTRAP]) 

years = ['2015', '2016', '2017', '2018', '2019', '2020']

data_pivot = data.pivot( index= ['GEO', 'Age group', 'Sex', 'Characteristics', 'REF_DATE'], columns= 'Indicators', values='VALUE').reset_index()
data_pivot['REF_DATE'] = data_pivot['REF_DATE'].astype(int)
indicatorList =  data_pivot.columns.tolist() #this works
indicatorList.remove('GEO')
indicatorList.remove('Age group')
indicatorList.remove('Sex')
indicatorList.remove('Characteristics')
indicatorList.remove('REF_DATE')

data_pivot = data.pivot( index= ['GEO', 'Indicators', 'Sex', 'Characteristics', 'REF_DATE'], columns= 'Age group', values='VALUE').reset_index()
data_pivot['REF_DATE'] = data_pivot['REF_DATE'].astype(int)
agegroupList =  data_pivot.columns.tolist()
agegroupList.remove('GEO')
agegroupList.remove('Indicators')
agegroupList.remove('Sex')
agegroupList.remove('Characteristics')
agegroupList.remove('REF_DATE')

data_pivot = data.pivot( index= ['GEO', 'Indicators', 'Age group', 'Characteristics', 'REF_DATE'], columns= 'Sex', values='VALUE').reset_index()
data_pivot['REF_DATE'] = data_pivot['REF_DATE'].astype(int)
sexList =  data_pivot.columns.tolist()
sexList.remove('GEO')
sexList.remove('Indicators')
sexList.remove('Age group')
sexList.remove('Characteristics')
sexList.remove('REF_DATE')

data_pivot = data.pivot( index= ['GEO', 'Indicators', 'Age group', 'Sex', 'REF_DATE'], columns= 'Characteristics', values='VALUE').reset_index()
data_pivot['REF_DATE'] = data_pivot['REF_DATE'].astype(int)
characteristicList =  data_pivot.columns.tolist()
characteristicList.remove('GEO')
characteristicList.remove('Indicators')
characteristicList.remove('Age group')
characteristicList.remove('Sex')
characteristicList.remove('REF_DATE')
            
grList = ['GEO', 'Region', 'Language'] #this is correct

# 3 Create the app's layout 

app.layout = html.Div([
    
    html.H1('Health Canada Indexes',
        style={'color': 'blue',
               'fontSize': '40px'}),
        html.H2('Health Canada'),
      
    ### Include the bar graph
    html.Br(),
    dcc.Dropdown( id='indicator', value='Arthritis (15 years and over)', 
                  options= [ {'label':ind, 'value':ind } for ind in  indicatorList] ),
    dcc.Dropdown( id='age', value='Total, 12 years and over', 
                  options= [ {'label':age, 'value':age } for age in  agegroupList] ),
    dcc.Dropdown( id='sex', value='Both sexes', 
                  options= [ {'label':sex, 'value':sex } for sex in  sexList] ),
    dcc.Dropdown( id='char', value='Number of persons', 
                  options= [ {'label':char, 'value':char } for char in  characteristicList] ),
    dcc.Dropdown( id='group_id', value='GEO', 
                  options= [ {'label':gr, 'value':gr } for gr in grList  ] ),
    dcc.Graph(id='id_chart'),

] )


@app.callback( Output(component_id='id_chart',component_property='figure'),
              Input(component_id='indicator', component_property='value'),
              Input(component_id='age', component_property='value'),
              Input(component_id='sex', component_property='value'),
              Input(component_id='char', component_property='value'),
              Input(component_id='group_id', component_property='value') )

def display_selected_pop_year(indicator, age, sex, char, grouper ):
    
    
    df = ((healthcan_df[(healthcan_df['Age group']==str(age))&(healthcan_df['Sex']==str(sex))
                   &(healthcan_df['Characteristics']==str(char))])).sort_values(indicator).dropna(subset=[indicator, grouper])   


    fig = px.scatter( df, x='REF_DATE', y=indicator,
            color=grouper,
            symbol=grouper,
            log_y=True,
            hover_name=df['GEO'],
            size=[1]* len(df),
            title= ' '.join([indicator, 'by', grouper, age, sex]),
            height=700,
            labels={"REF_DATE": "year"}
                    )
    
    return fig

app.run_server(mode='inline', height=600, width='100%', port=2456)
```
